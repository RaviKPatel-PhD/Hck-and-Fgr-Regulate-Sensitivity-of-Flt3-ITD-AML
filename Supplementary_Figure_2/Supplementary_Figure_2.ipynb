```python
collist = ['grey']*32
collist[1] = 'red'
collist
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,8)
ax.set(yscale='log')
#ax = sns.violinplot(x="Cancer Study", y="Value", data=Fgr, inner=None, scale = 'count', color = 'w')
ax = sns.swarmplot(y="Value", x="Cancer Study", palette= collist, size = 2, data=Fgr)
ax = sns.boxplot(x='Cancer Study', y= 'Value', data= Fgr, whis = 0, color = 'w', linewidth = 3, fliersize = 0)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90, size= 16)
ax.set(ylabel='Fgr Expression (FPKM)')
plt.setp(ax.artists, edgecolor = 'k', facecolor='w')
plt.setp(ax.lines, color='k')

;
```
